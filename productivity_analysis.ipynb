## High vs Low Productivity Comparison 

We compare habits on high prouctivity days vs low  productivity days.

```python
high = df[df["Productivity_Score"] >= 8]
low = df[df["Productivity_Score"] < 8]

comparison = pd.DataFrame({
    "High Productivity": high.mean(numeric_only=True),
    "Low Productivity": low.mean(numeric_only=True)
})

comparison.drop("Productivity_Score").plot(kind="bar", figsize=(8,5))
plt.title("High vs Low Productivity Habits")
plt.show()
```

## Productivity Level Classification

To simplify the analysis, productivity scores are grouped into three categories:

- High Productivity (Score ≥ 8)
- Medium Productivity (Score 5–7)
- Low Productivity (Score < 5)

This categorization helps identify behavioral differences between highly productive and less productive days.

```python
def categorize(score):
    if score >=8:
        return "High"
    elif score >=5:
        return "Medium"
    else:
         return "Low"
    df["Productivity_Level"]=df["Productivity_Score"].apply(categorize)
    df.head()
```

### Observation

Most days fall into the medium and high productivity categories. This suggests that consistent habits such as sufficient sleep, regular exercise, and focused study time contribute positively to overall productivity.
